```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
df.head()
```

```python
#Show DataFrame info
df.info()
```

```python
#Describe DataFrame
df.describe()
```

```python
df['Response'].value_counts()
```

```python
#Show a plot of the total number of responses
fig, ax = plt.subplots()
ax.bar(x = df['Response'].value_counts().index, height = df['Response'].value_counts().values)
ax.set_title("Total number of responses")
ax.set_xlabel("Response")
ax.set_ylabel("Count")
plt.show()
```

```python
#Show a plot of the response rate by the sales channel
df['Sales Channel'].value_counts()
```

```python
df[df['Response'] == 'Yes']
```

```python
df['Sales Channel'].unique()
```

```python
df.groupby('Sales Channel') ['Response'].value_counts(normalize=True).unstack()
```

```python
plt.figure(figsize=(9, 5))
sns.countplot(data=df, x='Sales Channel', hue='Response')
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#- Show a plot of the response rate by incom
df['Income'].value_counts()
```

```python
df.groupby('Income') ['Response'].value_counts(normalize=True).unstack()
```

```python
plt.hist(df[df['Response'] == 'Yes']['Income'], bins=50, alpha=0.5, label='Yes')
plt.hist(df[df['Response'] == 'No']['Income'], bins=50, alpha=0.5, label='No')
plt.title('Response rate by Income')
plt.ylabel('Number of Customers')
plt.xlabel('Income')
plt.legend(loc='upper right')
plt.show()
```

```python
#Categories income into groups :
bins = [0, 20000, 40000, 60000, 80000, 100000, 200000, float('inf')]
labels = ['0-20k', '20-40k', '40-60k', '60-80k', '80-100k', '100-200k', '200k+']
df['Income group'] = pd.cut(df['Income'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Income group', hue='Response')
plt.title('Response rate by Income')
plt.xlabel('Income group')
plt.ylabel('Response')
plt.xticks(rotation=45)

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
# -Do the distributions for different numerical columns look symmetrical? Compute the skewness for each, and add a comment with your findings.
skewness_values = {}

for col in numerical.columns:
    skewness = numerical[col].skew()
    skewness_values[col] = skewness

# Pour afficher les valeurs de skewness pour chaque colonne
for col, skewness in skewness_values.items():
    print(f"Skewness de la colonne {col}: {skewness}")
    
#add a comment with your findings :
# > For normally distributed data, skewness should be about 0.
# >> Column "Unamed" doesn't count.
# >> Column "Since Policy Inception" is the most symetrical.
# >> Columns "Income" an "Months Since Last Claim" have slight skewness
# >> The other columns have high skewness.
```
